```python
import pandas as pd
mobile = pd.read_csv('mobile.csv')
#1
def first_name(s):
    r = 0
    for i in s:
        if i != ' ':
            r += 1
        else:
            return(s[0:r])

#2
def set_company(s):
    match s:
        case '050':
            return 'Pelephone'
        case '051': 
            return 'wecom'
        case '052':
            return 'Cellcom'
        case '053':
            return'Hot Mobile'
        case '054':
            return 'Partner'

    
mobile['First Name'] = mobile['Customer Name'].apply(first_name)
mobile['Company'] = mobile['Phone Number'].apply(lambda x: x[:3])
mobile['Company'] = mobile['Company'].apply(set_company)

#3
print(mobile['Company'].value_counts())
print(mobile['Company'].unique())
print(mobile['Company'].nunique())

#4
mobile['User Type'] = mobile.apply(lambda row: 'Talker' if row['Talked Minutes'] > row['Internet Minutes'] else 'Surfer', axis=1)
import numpy as np
mobile['User Type'] = np.vectorize(lambda talker, internet: 'Talker' if talker > internet else 'Surfer')(mobile['Talked Minutes'], mobile['Internet Minutes'])

#5
print(mobile.sort_values(['Talked Minutes']))

#6
print(mobile.sort_values(['Talked Minutes', 'Internet Minutes']))

#7
print(mobile.nlargest(5,'Talked Minutes'))

#8 
print(mobile.nsmallest(3,'Internet Minutes'))

#9
print(mobile[mobile.duplicated()])
mobile.drop_duplicates()

#10
print(mobile.sample(n = 3)[['Customer Name', 'Phone Number']])
```
